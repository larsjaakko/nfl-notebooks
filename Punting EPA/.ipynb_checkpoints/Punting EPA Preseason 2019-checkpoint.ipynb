# NFL punting EPA analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.interpolate import make_interp_spline, BSpline

sns.set(style="whitegrid", 
        palette="pastel", 
        font='Circular Spotify Text', 
        font_scale = 1.1, 
        rc= {'grid.color': '.95', 'xtick.color': '.3', 'ytick.color': '.3'})
```

## Data

```python
df = pd.read_csv("../data/reg_pbp_2018.csv")
```

```python
df.head()
```

# Analysis!
## EPA per game per team

```python
teams_games = df[(df['play_type'] == 'punt') & (df['qtr'] <= 4)].groupby(['posteam', 'game_id'])['epa'].sum().reset_index()
teams_games.head()
```

```python
ranked = teams_games.groupby('posteam')['epa'].mean().reset_index().sort_values('epa', ascending=False).reset_index(drop=True)
ranked
```

```python
teams_games_wp = df[(df['play_type'] == 'punt') & (df['qtr'] <= 4)].groupby(['posteam', 'game_id'])['wpa'].sum().reset_index()
teams_games_wp.head()

ranked_wp = teams_games_wp.groupby('posteam')['wpa'].mean().reset_index().sort_values('wpa', ascending=False).reset_index(drop=True)
ranked_wp
```

```python

team_order = teams_games.groupby(['posteam']).agg(['mean']).reset_index()
team_order = team_order.sort_values(('epa', 'mean'), ascending=False)
team_order = team_order['posteam']

bar_data = teams_games.copy()

bar_data['posteam'] = bar_data['posteam'].astype("category")
bar_data['posteam'].cat.set_categories(team_order, inplace=True)
bar_data = bar_data.sort_values('posteam')
```

```python
f, ax = plt.subplots(figsize=(20, 6))
ax = sns.barplot(y='epa',
                 x='posteam', 
                 data=bar_data, 
                 palette=sns.cubehelix_palette(50, start=.5, rot=-.75),
                 errwidth=0.5,
                 errcolor='darkgray',
                 capsize=.2
                  )

#plt.xticks(rotation=90)

ax.set_ylabel('EPA per game')
ax.set_xlabel('\nTeam')
ax.set_title('EPA from punting per game in the 2018 reg. season (excluding OT)\n', fontsize=20, weight='bold')


sns.despine(left=True, right=True, bottom=True, top=True)

plt.savefig('output/' + 'epa_per_game' + '.png', format='png', dpi=300, bbox_inches = "tight")
```
